Fitting experimantal data

```python
import phlab
from  matplotlib import pyplot as plt
```

Single shot calculations for fixed input parameters :

```python
ws=phlab.rixs()
model = ws.model_single_osc(name = '1d')
model.color = 'r'
exp = ws.experiment(file = 'test_data.csv',name = 'test data')
```

```python
model.run()
```

```python
plt.figure(figsize = (10,5))
vitem=ws.visual(model_list=[model],exp=exp)
vitem.show(scale = 0)
```

Now we will get coupling constant from the fitting of the model to the test data. You can specidy any or all paramters from the input to be in the parameters space for the fitting. If you want to change fitting methods check [lmfit](https://lmfit.github.io/lmfit-py/fitting.html) page, by default it is 'brute'. 

```python
## fit model 2 experiment
model.param2fit.add(name = 'coupling', ivalue = model.input['coupling'] , range=[0.05,0.3])
model.fit(experiment = exp)
```

```python
model.run()
```

```python
plt.figure(figsize = (10,5))
vitem=ws.visual(model_list=[model],exp=exp)
vitem.show(scale = 0)
```
